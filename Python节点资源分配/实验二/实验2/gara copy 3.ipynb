# 资源分配问题

```python
import numpy as np
```

## 类定义
定义node，表示节点类

```python
class node():
    def __init__(self, name=None, resourse=None):
        self.name = name
        self.resourse = resourse # 节点所用资源的列表(在当前题目要求下，资源总和即可，不需要列表)
        self.fpga = None  # 节点被分配到的FPGA序号
        self.net = []     # 节点所属线网的序号列表
```

## 定义函数
### 文件信息读取
读取文件，取得节点资源信息

```python
def readNodeFile(fileName):
    nodes = {}
    f = open(fileName)
    for line in f:
        name = line.split()[0]
        resources = sum(map(int, line.split()[-10:]))
        nodes[name] = node(name, resources)
    f.close()
    return nodes
```

读取文件，取得线网信息

```python
def readNetFile(filename):
    lineStack = []
    nameList = []
    netId = 0
    f = open(filename)
    for line in f:
        lineStack.append(line.split())
    while lineStack != []:
        line = lineStack.pop()
        nameList.append(line[0])
        if 's' in line:
            for name in nameList:
                nodes[name].net.append(netId)
            nameList = []
            netId += 1
    f.close()
```

### 定义分配函数allocation

```python
def allocation():
    for node in nodes.values():
        node.fpga=np.random.randint(nosFPGA)
```

### 定义写文件函数writeResult

```python
def writeResult():
    for fpga in range(nosFPGA):
        f.write("F"+str(fpga))
        for node in nodes.values():
            if node.fpga==fpga:
                f.write(" "+node.name)
        f.write("\n")
```

### 定义资源差异函数 resourseCal
计算各个板的资源

```python
def resourseCal():
    resourses = [0]*nosFPGA
    for node in nodes.values():
        resourses[node.fpga]+=node.resourse
    return resourses
```

### 定义板间连线计算函数 linksCal
计算FPGA板间连线的总和

```python
def linksCal():
    links = 0
    for n1 in nodes.values():
        for n2 in nodes.values():
            if n1.net == n2.net and n1.fpga != n2.fpga:
                links += 1
    return links/2
```

### 清空分配信息
清除节点中的分配信息

```python
def clearAssignInfo():
    for node in nodes.values():
        node.fpga = None
```
